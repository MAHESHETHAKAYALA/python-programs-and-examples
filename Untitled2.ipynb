```python

print('welcome to ATM')

print('press 1 for deposite')
print('press 2 for withdraw')
print('press 3 for balance enquiry')
print('press 4 for exit')
while True:
  option=int(input('enter your option '))
  balance=100
  if(option==1):
    amount=float(input('enter amount to deposite'))
    if(amount%100==0):
      totalbalance=balance+amount
      print('amount has been deposited.......!')
      print('your total balance is {}'.format(totalbalance))
    else:
      print('invalid enters')
  elif(option==2):
    amount=float(input('enter amount to withdraw'))
    if(totalbalance>amount):
      totalbalance=totalbalance-amount
      print('withdraw success collect your cash')
      print('now your total balance is {}'.format(totalbalance))
    else:
      print('insufficient balance.....!')

  elif(option==3):
    print('your available balance is {}'.format(totalbalance))
  elif(option==4):
    print('thanks for visiting...have a good day........!')
    break
  else:
    print('invalid option choose the above')
```
