```python
def mult_yield():
  str1="siva"
  yield str1
  str2="mahesh"
  yield str2
  str3="sai vamsi sree ram prasad"
  yield str3
x=mult_yield()
print(next(x))
print(next(x))
print(next(x))

```

```python
class bank:
   def __init__(self,balance,deposit,withdrawl):
        self.balance=balance
        self.deposit=deposit
        self.withdrawl=withdrawl

   def balance_enquiry(self):
        print(self.balance)
   def withdrawl_amount(self,withdraw):
        print(self.balance-withdraw)
   def deposit_amount(self,deposit):
        print(self.balance+deposit)
balance=int(input("Enter balance amount "))
deposit=int(input("Enter deposit amount "))
withdraw=int(input("Enter withdraw amount "))
obj=bank(balance,deposit,withdraw)
obj.balance_enquiry(balance)
obj.deposit_amount(deposit)
obj.withdraw_amount(withdraw)







```
