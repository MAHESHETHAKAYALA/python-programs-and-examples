```python
num=int(input())
power=len(str(num))
#print(power)
sum=0
number=num
while num>0:
  digit=num%10
  sum=sum+digit**power
  num//=10

if sum==number:
  print(number,'is a armstrong number')
else:
  print(number ,'is not a armstrong number')
```
